```python
# Import necessary dependencies
import pandas as pd
import numpy as np
import pickle
```

```python
# Read pickle dataset and modify
enron_dataset = pd.read_pickle('Resources/final_project_dataset.pkl')
modified_dataset= pd.read_pickle('Resfinal_project_dataset_modified.pkl')
len(modified_dataset)
```

```python
# Insert a break for each of the 143 entries
from pprint import pprint
c = 0
for i in modified_dataset.items():
    if c < 143:
        c += 1
        pprint(i)
        print('\n')
    else:
        break
```
